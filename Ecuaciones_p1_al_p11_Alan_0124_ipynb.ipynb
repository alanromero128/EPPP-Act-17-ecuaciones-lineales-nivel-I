* Problema 1: $x + 8 = 15 $
* Planteamiento: Un número desconocido sumado a 8 da 15.
* Despeje: El 8 que suma pasa restando: $x = 15 - 8$.Solución: $x = 7$

```python
# --- Problema 1 ---
# x + 8 = 15
# resultado_total = 15
# numero_sumado = 8
resultado_total = int(input("Ingrese el resultado total: "))
numero_sumado = int(input("Ingrese el número sumado: "))
x = resultado_total - numero_sumado
print(f"Resultado Problema 1: x = {x}")
```

 Problema 3: $3x + 4x = 21$
* Planteamiento: Tres veces un número más cuatro veces el mismo número es 21.Simplificación: Sumamos términos semejantes: $(3+4)x = 21 \rightarrow 7x = 21$.Despeje: El 7 que multiplica pasa dividiendo: $x = 21 / 7$.Solución: $x = 3$.

```python
total = int(input("Introduce el resultado de 3x + 4x (ej. 21): "))

# Simplificamos 3x + 4x a 7x. Despejamos: x = total / 7
x = total / 7

print(f"El valor de x es: {x}")
```

Problema 5: $6a - 2a + a = 20$
* Planteamiento: Se combinan sumas y restas de la misma variable $a$.
* Simplificación: $6a - 2a = 4a$; luego $4a + 1a = 5a$. Entonces: $5a = 20$.
* Despeje: El 5 pasa dividiendo: $a = 20 / 5$.Solución: $a = 4$.

```python
print("--- Calculador de Dimensiones del Terreno ---")

# Pedimos el perímetro total
perimetro = int(input("Introduce el perímetro total del terreno (ej. 50): "))

# Pedimos la diferencia entre largo y ancho
diferencia = int(input("¿Cuántos metros más mide el largo que el ancho? (ej. 5): "))

# Aplicamos la fórmula despejada:
# El 2*diferencia es porque hay dos lados largos.
ancho = (perimetro - (2 * diferencia)) / 4
largo = ancho + diferencia

print(f"\nResultados:")
print(f"El ancho del terreno es: {ancho} metros")
print(f"El largo del terreno es: {largo} metros")
```

Problema 7: $4x + 3 = 2x + 11$
* Planteamiento: Hay variables y números en ambos lados.
* Agrupación: Pasamos las $x$ a la izquierda y los números a la derecha.$4x - 2x = 11 - 3$
* Simplificación: $2x = 8$.Despeje: $x = 8 / 2$.Solución: $x = 4$.

```python
print("Resolvamos 4x + B = 2x + D")
b = int(input("Introduce el valor de B (ej. 3): "))
d = int(input("Introduce el valor de D (ej. 11): "))

# Agrupamos: 4x - 2x = D - B -> 2x = D - B
x = (d - b) / (4 - 2)

print(f"Para la ecuación 4x + {b} = 2x + {d}, el valor de x es: {x}")
```

Problema 9: $9m + 4 = 5m + 20$
* Agrupación: $9m - 5m = 20 - 4$.
* Simplificación: $4m = 16$.
* Despeje: $m = 16 / 4$.
* Solución: $m = 4$.

```python
print("Resolvamos 9m + 4 = 5m + Resultado")
resultado = int(input("Introduce el valor final (ej. 20): "))

# Despeje: (9 - 5)m = resultado - 4
m = (resultado - 4) / (9 - 5)

print(f"El valor de m es: {m}")
```

Problema 11: $6a + 5 = 2a + 21$
* Agrupación: $6a - 2a = 21 - 5$.
* Simplificación: $4a = 16$.
* Despeje: $a = 16 / 4$.
* Solución: $a = 4$.

```python
# --- Problema 11 ---
print("Resolvamos la ecuación: 6a + 5 = 2a + Resultado")

# Pedimos el valor del lado derecho al usuario
resultado_derecho = int(input("Introduce el valor de la derecha (ej. 21): "))

# Despejamos siguiendo la lógica: (6-2)a = resultado_derecho - 5
# 4a = resultado_derecho - 5
a = (resultado_derecho - 5) / (6 - 2)

print(f"Para el resultado {resultado_derecho}, el valor de 'a' es: {a}")
```
