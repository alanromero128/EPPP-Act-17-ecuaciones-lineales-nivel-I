* Problema 1: $x + 8 = 15 $
* Planteamiento: Un número desconocido sumado a 8 da 15.
* Despeje: El 8 que suma pasa restando: $x = 15 - 8$.Solución: $x = 7$

```python
# --- Problema 1 ---
# x + 8 = 15
resultado_p1 = 15 - 8
print(f"Resultado Problema 1: x = {resultado_p1}")
```

 Problema 3: $3x + 4x = 21$
* Planteamiento: Tres veces un número más cuatro veces el mismo número es 21.Simplificación: Sumamos términos semejantes: $(3+4)x = 21 \rightarrow 7x = 21$.Despeje: El 7 que multiplica pasa dividiendo: $x = 21 / 7$.Solución: $x = 3$.

```python
# --- Problema 3 ---
# 3x + 4x = 21 -> 7x = 21
x_p3 = 21 / (3 + 4)
print(f"Resultado Problema 3: x = {int(x_p3)}")
```

Problema 5: $6a - 2a + a = 20$
* Planteamiento: Se combinan sumas y restas de la misma variable $a$.
* Simplificación: $6a - 2a = 4a$; luego $4a + 1a = 5a$. Entonces: $5a = 20$.
* Despeje: El 5 pasa dividiendo: $a = 20 / 5$.Solución: $a = 4$.

```python
# --- Problema 5 ---
# 6a - 2a + a = 20 -> 5a = 20
a_p5 = 20 / (6 - 2 + 1)
print(f"Resultado Problema 5: a = {int(a_p5)}")
```

Problema 7: $4x + 3 = 2x + 11$
* Planteamiento: Hay variables y números en ambos lados.
* Agrupación: Pasamos las $x$ a la izquierda y los números a la derecha.$4x - 2x = 11 - 3$
* Simplificación: $2x = 8$.Despeje: $x = 8 / 2$.Solución: $x = 4$.

```python
# --- Problema 7 ---
# 4x + 3 = 2x + 11 -> (4x - 2x) = (11 - 3)
x_p7 = (11 - 3) / (4 - 2)
print(f"Resultado Problema 7: x = {int(x_p7)}")
```

Problema 9: $9m + 4 = 5m + 20$
* Agrupación: $9m - 5m = 20 - 4$.
* Simplificación: $4m = 16$.
* Despeje: $m = 16 / 4$.
* Solución: $m = 4$.

```python
# --- Problema 9 ---
# 9m + 4 = 5m + 20 -> (9m - 5m) = (20 - 4)
m_p9 = (20 - 4) / (9 - 5)
print(f"Resultado Problema 9: m = {int(m_p9)}")
```
